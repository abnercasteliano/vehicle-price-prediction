```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importando os dados
df = pd.read_csv('cardata.csv')
```

```python
df.head()
```

```python
# verificando o tamanho do dataset
df.shape
```

```python
# vefificando o tipo dos dados
df.dtypes
```

```python
# Verificando os valores das variáveis categóricas mais relevantes

print('Fuel_Type_labels: {0}'.format(df['Fuel_Type'].unique()))
print('Seller_Type_labels: {0}'.format(df['Seller_Type'].unique()))
print('Transmission_labels: {0}'.format(df['Transmission'].unique()))
```

```python
# Verificando missing Values

df.isnull().sum()
```

```python
# Eliminando as variáveis indesejadas

new_dataframe = df.drop(['Car_Name'], axis=1)
```

```python
new_dataframe.head()
```

## Transformando a data do ano para número de anos

```python
# criando a variável Ano atual

new_dataframe['Current_Year'] = 2021
```

```python
new_dataframe['Num_years'] = new_dataframe['Current_Year'] - new_dataframe['Year']

new_dataframe.head()
```

```python
# eliminando Year e Current_years

new_dataframe.drop(['Year', 'Current_Year'], axis=1, inplace=True)

new_dataframe.head()
```

## Transformando a Variáveis categóricas em números

```python
# aplicando one hot encoder
new_dataframe = pd.get_dummies(new_dataframe, drop_first=True)
```

```python
new_dataframe.head()
```

## Analisando as variáveis

```python
# analisando as correlações

def plot_corr(df):
    plt.figure(figsize=(15, 9))
    corrmat = df.corr()
    top_corr = corrmat.index
    sns.heatmap(df[top_corr].corr(),
            annot = True,
            fmt = '.2f',
            cmap='RdYlGn')
    plt.title('Correlação entre variáveis do dataset')
    plt.show()

    
plot_corr(new_dataframe)
```

### Através do gráfico podemos concluir que trata-se de um modelo de regressão linear

## Dividindo e treinando o modelo

```python
# dividindo os dados em INPUT e OUTPUT

X = new_dataframe.drop(['Selling_Price'], axis=1)

y = new_dataframe['Selling_Price']
```

```python
# Escolha das variáveis mais importantes com extraTreeRegressor

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X, y)
```

```python
# plotando um gráfico de feature importances

feature_importances = pd.Series(model.feature_importances_, index=new_dataframe.drop(['Selling_Price'], axis=1).columns)
feature_importances.nlargest(5).plot(kind='barh')
plt.show()
```

## Normalização dos dados com MinMaxScaler

```python
# importando MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scale = scaler.fit(X)
X = scale.transform(X)
```

## Treinando os dados com Linear Regression

```python
# Dividindo os dados em Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# importando o Linear Regression

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
# treinando o modelo
lr_model.fit(X_train, y_train.ravel())

# Validação do modelo
Y_pred = lr_model.predict(X_test)
```

```python
# plotando os resultados

sns.set(rc={'figure.figsize':(11,8)})
sns.regplot(x=y_test, y=Y_pred)
plt.show()
```

```python
# avaliando o score do modelo

from sklearn.metrics import r2_score

print('Score: {0:.2f}%'.format(r2_score(y_test, Y_pred) * 100))
```

## Otimizando o modelo com Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
# Definição de hiperparâmetros

# número de árvores de 100 a 1200
n_estimators= [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
# máximo de variáveis por split do nó
max_features = ['auto', 'sqrt']
# número máximo de níveis da árvore
max_depth = [int(x) for x in np.linspace(start=5, stop=30, num=6)]
# número máximo de amostras por split
min_samples_split = [2, 5, 10, 15, 100]
# número minimo de amostras por folha
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Criando uma grade aleatória de hiperparâmetros

random_grid = {'n_estimators' : n_estimators,
              'max_features' : max_features, 
              'max_depth' : max_depth, 
              'min_samples_split' : min_samples_split,
              'min_samples_leaf' : min_samples_leaf}

print(random_grid)
```

```python
# Optimização dos hiperparâmetros com RandomizedSearchCV

from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(estimator = rf_regressor, 
                                   param_distributions = random_grid,
                                   scoring = 'r2',
                                   n_iter = 10,
                                   cv = kfold,
                                   verbose = 2,
                                   random_state=None,
                                   n_jobs = 1)
```

```python
# Treinando o modelo utilizando a otimização

random_search.fit(X_train, y_train.ravel())
```

```python
# Analisando os resultados

Y_prediction = random_search.predict(X_test)
```

```python
Y_prediction
```

```python
# plotando os resultados

sns.set(rc={'figure.figsize':(11,8)})
sns.regplot(x=y_test, y=Y_prediction)
plt.show()
```

```python
from sklearn.metrics import r2_score

print('Score: {0:.2f}%'.format(r2_score(y_test, Y_prediction) * 100))
```

## Salvando o modelo no formato pickle com joblib

```python
import joblib

joblib.dump(random_search, 'RandomFRegressor.pkl')
```
